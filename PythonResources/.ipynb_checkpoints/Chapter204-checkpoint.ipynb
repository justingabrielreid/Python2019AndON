```python
cities, years = [],[]

for game in open('games.txt','r'):
    words = game.split()
    
    city = ' '.join(words[:-1])
    year = words[-1].strip('()')
    
    cities.append(city)
    years.append(year)
    
```

Geolocated olympic-city cooordinates from a JSON file

```python
coordinates_by_city = json.load(open('coords.json', 'r'))
coordinates_by_city
```

```python
postWW2OlympicGames = []
for city,year in zip(cities,years):
    if int(year) > 1945:
        postWW2OlympicGames.append(city + " : " + year)

postWW2OlympicGames
        
```

```python
#creating a dictionary

cities_by_year = {year: city for city, year in zip(cities, years)}
```

```python
cities_by_year
```

```python
#doing the long way to see if I understand it. 

cities_by_year2 = {}
for year,city in zip(years,cities):
    cities_by_year2[year] = city
```

```python
cities_by_year2
```

```python
#set defintion
cities_after_1930 = {city for year, city in cities_by_year.items() if int(year) > 1930}
```

```python
cities_after_1930
```

```python
#Long Way of doing this

cities_after_1940 = set()
for year,city in cities_by_year.items():
    if int(year) > 1940:
        cities_after_1940.add(city)
```

```python
cities_after_1940
```

```python
#creating a map
pp.figure(figsize=(8,8))
#creating a basemap
world=Basemap(projection='ortho',lat_0=75,lon_0=0)
#drawlines
world.drawcoastlines(linewidth=0.25)
world.drawcountries(linewidth=0.25)

for year, city in cities_by_year.items():
    x,y = world(*coordinates_by_city[city])
    
    world.plot(x,y,'r.')
    pp.text(x,y,year,fontsize=10,ha='center',va='bottom',color='navy')
```

```python
#Writing a generator that will compute a fib sequence
#Will never terminate

def fibonacci():
    f1, f2 = 0, 1
    while True:
        yield f2
        f1 = f2
        f2 = f1 + f2
        
```

```python
f = fibonacci()
```

```python
[next(f) for index in range(20)] 
# next(f) is shorthand for the next method. f.__next__()
#This will generate the first twenty fib numbers
```
